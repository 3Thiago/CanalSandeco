## RECONHECENDO NÚMERO ESCRITO A MÃO

```python
from tensorflow.keras.preprocessing.image import load_img

import numpy as np
```

```python
img = load_img('8.png', grayscale=True)

print(img.size)
```

## MOSTRANDO A IMAGEM

```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.imshow(img)
plt.show(img)
```

## INVERTENDO A IMAGEM

```python
import PIL.ImageOps

img = PIL.ImageOps.invert(img)
```

```python
plt.imshow(img)
plt.show(img)
```

### CONVERTENDO IMAGEM EM ARRAY

```python
from tensorflow.keras.preprocessing.image import img_to_array
```

```python
img_array = img_to_array(img)
```

```python
img_array
```

```python
img_array.shape
```

## FLATTENING

```python
img_array = img_array.reshape((1,28*28))
img_array.shape
```

## NORMALIZANDO

```python
img_array = img_array.astype('float32')/255
```

```python
img_array
```

## CARREGANDO O MODELO

```python
from tensorflow.keras.models import load_model
```

```python
model = load_model('mnist-t0.11.h5')
```

```python
preds = model.predict_classes(img_array)
```
